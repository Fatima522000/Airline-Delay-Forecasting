#Preprocessing

#1.Cleaning

```python
#missing_values
from sklearn.impute import KNNImputer
airport_col = df['airport']
carrier_col = df['carrier']
df=df.drop(columns=['airport', 'carrier'])

knn_imputer = KNNImputer(n_neighbors=5)
df = pd.DataFrame(knn_imputer.fit_transform(df), columns=df.columns)
df['airport'] = airport_col
df['carrier'] = carrier_col

df.isnull().sum()
```

```python
#smooth_outliers
columns_to_fix = ['arr_flights', 'arr_del15', 'carrier_ct', 'weather_ct',
                  'nas_ct', 'late_aircraft_ct', 'arr_delay', 'carrier_delay',
                  'weather_delay', 'nas_delay', 'late_aircraft_delay']

for column in columns_to_fix:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR
    median_value = df[column].median()
    df[column] = df[column].apply(lambda x:median_value if x<lower_bound or x>upper_bound else x)
```

```python
df.describe()
```

#2.Transformation

```python
#normalization  using z-score (Mean Absolut deviatin)
columns_to_normalize = ['arr_flights', 'arr_del15', 'carrier_ct', 'weather_ct',
                        'nas_ct', 'late_aircraft_ct', 'arr_delay', 'carrier_delay',
                        'weather_delay', 'nas_delay', 'late_aircraft_delay']

for column in columns_to_normalize:
    mean_value = df[column].mean()
    mad_value = df[column].apply(lambda x: abs(x - mean_value)).mean()
    df[column] = (df[column] - mean_value) / mad_value
```

```python
plt.figure(figsize=(16, 5))
sns.boxplot(df.iloc[:, 2:])
plt.title("Boxplot after Standardization", fontsize=14)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Standardized Values", fontsize=12)

plt.show()

```

#3.Dimensionality Reduction

```python
df['airport_carrier']=df['airport']+"_"+df['carrier']
df.drop(columns=['airport','carrier'],inplace=True)
```

```python
from sklearn.decomposition import PCA
columns_to_pca=[ 'arr_delay', 'carrier_delay',
                  'weather_delay', 'nas_delay', 'late_aircraft_delay']
X=df[columns_to_pca]
pca=PCA(n_components=2)
principal_components=pca.fit_transform(X)
pca_df=pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df = df.drop(columns=columns_to_pca)
df = pd.concat([df, pca_df], axis=1)
```

##Train and Test split

```python
#ARIMA Algorithm
series_1=df['arr_del15']
split_point=int(len(series_1)*0.8)
train=series_1.iloc[:split_point]
test=series_1.iloc[split_point:]
```

```python
#ARIMAX
series_2=df['arr_del15']
exogenous_features=df.drop(columns=['arr_del15'])
train_exog=exogenous_features.iloc[:split_point]
test_exog=exogenous_features.iloc[split_point:]
train_series=series_2.iloc[:split_point]
test_series=series_2.iloc[split_point:]
```

```python
#LightGBM (Light Gradient Boosting Machine)
from sklearn.model_selection import train_test_split
series_3=df['arr_del15']
exogenous_features=df.drop(columns=['arr_del15'])

X_train,X_test, y_train, y_test=train_test_split(exogenous_features,series_3, test_size=0.2, shuffle=False)
```

#Models and Evaluate

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def evaluate(y_test,forecast ):
  MAE=mean_absolute_error(y_test,forecast )
  MSE=mean_squared_error(y_test,forecast )
  RMSE=np.sqrt(MSE)

  print('MAE:',MAE,'MSE:',MSE,'RMSE:',RMSE)
```

```python
#ARIMA
from statsmodels.tsa.arima.model import ARIMA

model_1=ARIMA(train,order=(0,0,0))
model_fit=model_1.fit()

forecast=model_fit.forecast(steps=len(test))
```

```python
evaluate(test,forecast)
```

```python
#ARIMAX
model_arimax = ARIMA(train_series, exog=train_exog, order=(0,0,0))
model_fit_arimax = model_arimax.fit()
forecast_arimax = model_fit_arimax.predict(start=len(train_series), end=len(series_2)-1, exog=test_exog)
```

```python
evaluate(test_series,forecast_arimax)
```

```python
#LightGBM
from lightgbm import LGBMRegressor
model_lgbm = LGBMRegressor(random_state=42)
model_lgbm.fit(X_train, y_train)
forecast_lgbm=model_lgbm.predict(X_test)
```

```python
evaluate(y_test,forecast_lgbm)
```
